#EXPERIMENTO 4

```python
i=np.array([0.052,0.256,0.505,0.752,1.026,1.244])
b=np.array([0.57,0.72,0.87,1.05,1.24,1.36])

x=i
y=b
n = len(x)
sumax = sum(x)
sumay = sum(y)
sumax2 = sum(x*x)
sumay2 = sum(y*y)
sumaxy = sum(x*y)
promx = sumax/n
promy = sumay/n
B = (n*(sumaxy) - sumax*sumay)/(n*sumax2 - pow(sumax,2))
A = (sumax2*sumay - sumax*sumaxy)/(n*sumax2 -pow(sumax,2))


fig, axes=plt.subplots(1,1)
axes.plot(i,b,'.',label='Datos')
axes.plot(i,A+B*i,'-',label='$linear    fit$')
axes.set_title('Curva de calibracin para d=R y z=0')
axes.set_xlabel('i/A')
axes.set_ylabel('Bh/mT')
axes.legend()
```
